<span style="color:#C70039;font-size:25pt;"> <b> 1.Business Case      </b>  </span>

Bike sharing systems are a new version of traditional bike rentals where the whole process from membership, rental and return back has been automated. Through these systems, users are easily able to rent a bike from a particular position and return back at another position. Currently, there are about over 500 bike-sharing programs around the world which is composed of over 500 thousands bicycles. Today, there exists great interest in these systems due to their important role in traffic, environmental and health issues. Apart from interesting real world applications of bike sharing systems, the characteristics of data being generated by these systems make them attractive for research. Opposed to other transport services such as bus or subway, the duration of travel, departure and arrival position is explicitly recorded in these systems. This feature turns the bike sharing system into a virtual sensor network that can be used for sensing mobility in the city. Hence, it is expected that most of the important events in the city could be detected via monitoring these data.

##  Common function that will print all the scores for each model.

```python

def Evaluate(model, Y_Test, y_train, y_Pred_test,y_Pred_train) :
    
    model_name = model
    
    # Number of observations is the shape along axis 0.
    n = X_test.shape[0]
    
    # Number of features (predictors, p) is the shape along axis 1.
    p = X_test.shape[1]    
    
    # r2 score for train data.
    Train_acc_score = r2_score(y_train, y_Pred_train)
    
    # r2 score for test data.
    Test_acc_score = r2_score(y_test, y_Pred_test)
    
    # Root Mean Square error for train data.
    RMSE_train = np.sqrt(mean_squared_error(y_train, y_Pred_train))
    
    # Root Mean Square error for test data.
    RMSE_test = np.sqrt(mean_squared_error(y_test, y_Pred_test))
   
    # Adjusted R-squared formula for train
    Train_Adjusted_r2_score = 1-(1-Train_acc_score)*(n-1)/(n-p-1)
    
    # Adjusted R-squared formula for test
    Test_Adjusted_r2_score = 1-(1-Test_acc_score)*(n-1)/(n-p-1)
    
    # Method to calculte the mean_absolute_error for train.
    MAE_train = mean_absolute_error(y_Pred_train,y_train)
    
    # Method to calculte the mean_absolute_error for test.
    MAE_test = mean_absolute_error(y_Pred_test,y_test)
    
     # Method to calculte the mean_squared_error for train.
    MSE_train = mean_squared_error(y_Pred_train,y_train)
    
    # Method to calculte the mean_squared_error for test.
    MSE_test = mean_squared_error(y_Pred_test,y_test)
    
    
    Metric = {'Train Acc_Score' : f'{Train_acc_score * 100:.0f}%', 
        
              'Test Acc_Score' : f'{Test_acc_score * 100:.0f}%',
              
              'RMSE_train' : format(RMSE_train ,'.3g'),
              
              'RMSE_test' : format(RMSE_test,'.3g'),
              
              'Train_Adjust_r2_score' : round(Train_Adjusted_r2_score ,3),
              
              'Test_Adjust_r2_score' : round(Test_Adjusted_r2_score ,3),
              
              'MAE_train' : format(MAE_train,'.3g'),
              
              'MAE_test' : format(MAE_test,'.3g'),
              
              'MSE_train' : format(MSE_train,'.3g'),
              
              'MSE_test' : format(MSE_test,'.3g')              
   
             }
    

    print( "  \n   ")
    
    print('*'*90)
    
    print("                                      " , model_name)
    
    print('*'*90)
    
    print( "  \n   ")
    
    print(f'TRAIN Acc Score : {Train_acc_score * 100:.2f}%')
    
    print(f'TEST Acc Score : {Test_acc_score * 100:.2f}%')
   
    print('='*65)
  
    print('RMSE - TRAINING : ', RMSE_train)  #check root mean squared error 
    print('RMSE - TESTING : ', RMSE_test)

    print('='*65)
    
    print('Train_Adjusted_r2_score : ', Train_Adjusted_r2_score)
    print('Test_Adjusted_r2_score : ', Test_Adjusted_r2_score)
    
    print('='*65)
   
    print("mean_absolute_error - TRAINING : " , MAE_train)
    print("mean_absolute_error - TESTING : " , MAE_test)

    print('='*65)
   
    print("mean_squared_error - TRAINING : " , MSE_train )
    print("mean_squared_error - TESTING : " , MSE_test)    
    print( "  \n   ")
    print('*'*90)
    
    plt.figure(figsize=(4,4))
    plt.scatter(y_test, y_Pred_test, c='crimson')
    plt.yscale('log')
    plt.xscale('log')
    
    p1 = max(max(y_Pred_test), max(y_test))
    p2 = min(min(y_Pred_test), min(y_test))
    plt.plot([p1, p2], [p1, p2], 'b-')    
    plt.xlabel('Actual Values', fontsize=15)
    plt.ylabel('Predictions', fontsize=15)
    plt.axis('equal')
    plt.show()
    
    return Metric
```
