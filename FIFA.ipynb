```python
for index,row in df.iterrows():
    if row['st']>max(row['lw'],row['rw']):
        df.loc[index,'player_positions']='st'
    elif row['preferred_foot']==0:
            df.loc['index','player_positions']='lw'
    else:
         df.loc['index','player_positions']='rw'
mean_eur_rw=(df.loc[df['player_positions']=='RW','wage_eur'].mean())
mean_eur_lw=(df.loc[df['player_positions']=='LW','wage_eur'].mean())
mean_eur_st=(df.loc[df['player_positions']=='st','wage_eur'].mean())
print('The  average wage of a Striker',mean_eur_st)
print('The  average wage of a Right Winger',mean_eur_rw)
print('The  average wage of a Left Winger',mean_eur_lw)
```

<span style="color:limegreen;font-size:20pt;font-family: Arial Black;"> <i> BUILDING MODEL </i> </span><br>

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2,random_state=10)

model.fit(X)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
from scipy.spatial.distance import cdist
wcss=[]

for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # inertia is nothing but sum of squares
```

```python
wcss
```

```python
plt.plot(range(2,12),wcss,marker='d',)
plt.title('Elbow method')
plt.xlabel('No of clusters(K)')
plt.ylabel('WCSS')
plt.show()
```

```python
pred=model.predict(X)
```

```python
labels = model.labels_
labels
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X,labels)
print(f'K-Means Clustering Silhouette score: {score:.2f}')
```
